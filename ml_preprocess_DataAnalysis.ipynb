```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('df4.csv', encoding='ISO-8859-1')
df.head()
```

```python
df['sentiment'][df['sentiment'] != -1].value_counts().plot.pie()
```

```python
#On remarque que notre dataset is unbalanced
class_counts = df['sentiment'].value_counts()

# Affichage des r√©sultats
print(class_counts)
```

```python
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(nrows = 1,ncols = 2)
fig.set_size_inches(15, 7.5)
                   
sns.countplot(data=df, x="sentiment",ax=ax1)
df["sentiment"].value_counts().plot.pie(labels = ["negative","positive", "neutral"], autopct="%0.2f", ax=ax2)

ax1.set_title("class counts by label (Bar plot)")
ax2.set_title("class counts by label (Pie plot)")

```
